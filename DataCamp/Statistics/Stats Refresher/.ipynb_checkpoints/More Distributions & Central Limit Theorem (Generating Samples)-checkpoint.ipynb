## More Distributions & Central Limit Theorem

Since each deal Amir worked on (both won and lost) was different, each was worth a different amount of money. These values are stored in the amount column of amir_deals As part of Amir's performance review, you want to be able to estimate the probability of him selling different amounts, but before you can do this, you'll need to determine what kind of distribution the amount variable follows.

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
amir_deals = pd.read_csv('../Datasets/amir_deals.csv', index_col=0)
amir_deals.head()
```

```python
sns.histplot(amir_deals['amount'], bins=10)
plt.show()
```

```python
from scipy.stats import norm
print(help(norm.cdf))
```

### Probabilities from the Normal Distribution

* We can answer this using norm-dot-cdf from scipy-dot-stats, 
    * which takes the area of the normal distribution less than some number. 


```python
# Probability of a deal < 7500

# Function takes the value of interest (area less than), the normal distributions' mean, and standard deviation
prob_less_7500 = norm.cdf(7500, amir_deals['amount'].mean(), amir_deals['amount'].std())
print("The probability Amir's deal was less than $7,500 was : {}".format(prob_less_7500))

# Probability of deal > 1000
prob_over_1000 = 1 - norm.cdf(1000, 5000, 2000)
print("The probability Amir's deal was more than $1,000 was : {}".format(prob_over_1000))

# Probability of deal between 3000 and 7000
prob_3000_to_7000 = norm.cdf(7000, amir_deals['amount'].mean(), amir_deals['amount'].std()) - norm.cdf(
                            3000, amir_deals['amount'].mean(), amir_deals['amount'].std())
print("Probability of deal between these amounts (3,000 - 7,000) is : {}".format(prob_3000_to_7000))

# What amount will 25% of Amir's sales be less than
amir_25_deal_whisker = norm.ppf(.25, amir_deals['amount'].mean(), amir_deals['amount'].std())
print("25% of Amir's sales will be less than : {}".format(amir_25_deal_whisker))
```

### Simulating Sales Under New Market Conditions
The company's financial analyst is predicting that next quarter, the worth of each sale will increase by 20% and the volatility, or standard deviation, of each sale's worth will increase by 30%. To see what Amir's sales might look like next quarter under these new market conditions, you'll simulate new sales amounts using the normal distribution and store these in the new_sales DataFrame, which has already been created for you.

```python
# Calculate new average amount = (mean * percent increase) + mean
new_mean = (5000 * .2) + 5000

# Calculate new standard deviation
new_sd = (2000 * .3) + 2000

# Simulate 36 new sales
new_sales = norm.rvs(new_mean, new_sd, size=36)

# # Create histogram and show
sns.histplot(new_sales)
plt.show()
```

```python
plt.hist(new_sales)
plt.show()
```

### Poisson Distribution


#### Tracking Lead Responses
Your company uses sales software to keep track of new sales leads. It organizes them into a queue so that anyone can follow up on one when they have a bit of free time. Since the number of lead responses is a countable outcome over a period of time, this scenario corresponds to a Poisson distribution. On average, Amir responds to 4 leads each day. In this exercise, you'll calculate probabilities of Amir responding to different numbers of leads.

```python
# Import poisson from scipy.stats
from scipy.stats import poisson

# Probability of 5 responses, given that he responds to an average of 4
prob_5 = poisson.pmf(5,4)

print(prob_5)

# Amir's coworker responds to an average of 5.5 leads per day. 
# What is the probability that she answers 5 leads in a day?

# Probability of 5 responses
prob_coworker = poisson.pmf(5, 5.5)

print(prob_coworker)

# What's the probability that Amir responds to 2 or fewer leads in a day?
# Probability of 2 or fewer responses
prob_2_or_less = poisson.cdf(2, 4)

print(prob_2_or_less)

# What's the probability that Amir responds to more than 10 leads in a day?
# Probability of > 10 responses
prob_over_10 = 1 - poisson.cdf(10, 4)

print(prob_over_10)
```

#### Exponential Exercises
* To further evaluate Amir's performance, you want to know how much time it takes him to respond to a lead after he opens it. On average, it takes 2.5 hours for him to respond. In this exercise, you'll calculate probabilities of different amounts of time passing between Amir receiving a lead and sending a response.
* recall `expon.cdf()` which takes the average value in the scale and return a less than probability which can be inversed simply by taking away one from the decimal percentage value from the method return

```python
# Import expon from scipy.stats
from scipy.stats import expon

# Print probability response takes < 1 hour
print(expon.cdf(1, scale=2.5))

# Print probability response takes > 4 hours
print(1 - expon.cdf(4, scale=2.5))

# Print probability response takes 3-4 hours (Interval is the highest minus the low level to find the in between range)
print(expon.cdf(4, scale=2.5) - expon.cdf(3, scale=2.5))
```
